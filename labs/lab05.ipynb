Імпортуємо необхідні біліотеки

```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import animation, rc
```

Налаштуємо параметри відображення анімацій.

```python
# equivalent to rcParams['animation.html'] = 'html5'
rc('animation', html='html5')
```

Задамо параметри траєкторії:

- розмір квадрату (довжина ребра) $$s=7$$
- центр квадрату $$(x_c, y_c) = (5, 5)$$

```python
SQUARE_SIZE = 4
XC, YC = 6, 6
```

Розрахуємо «границі» квадрату:

$x_{left} = x_c - \frac{s}{2}$ , $x_{right} = x_c + \frac{s}{2}$;

$y_{top} = y_c + \frac{s}{2}$, $y_{bottom} = y_c - \frac{s}{2}$

```python
X_LEFT = XC - SQUARE_SIZE / 2
X_RIGHT = XC + SQUARE_SIZE / 2

Y_BOTTOM = YC - SQUARE_SIZE / 2
Y_TOP = YC + SQUARE_SIZE / 2
```

```python
FRAMES = 60
STEP = SQUARE_SIZE / (FRAMES / 4)
```

Побудуємо початкове зображення

```python
fig = plt.figure()
fig.set_dpi(100)

ax = plt.axes(xlim=(0, 12), ylim=(0, 12))
path_data = (
    (X_LEFT, Y_TOP),
    (X_LEFT, Y_BOTTOM),
    (X_RIGHT, Y_BOTTOM),
    (X_RIGHT, Y_TOP),
    (X_LEFT, Y_TOP),
)
_x, _y = zip(*path_data)
line, = ax.plot(_x, _y, 'go-')

obj = plt.Circle((XC, YC), 1, fc='y')


ax.grid()
ax.set_aspect('equal', 'datalim')
```

Вкажемо, як відображувати об'єкт у початковій точці траекторії:

```python
def init():
    obj.center = (
        SQUARE_SIZE / 2 + XC, 
        SQUARE_SIZE / 2 + YC, 
    )
    ax.add_patch(obj)
    return obj,
```

Опишемо траекторію руху:

```python
def animate(i):
    x, y = obj.center
    if i  < 0.25 * FRAMES:
        x -= STEP
    elif i < 0.5 * FRAMES:
        y -= STEP
    elif i < 0.75 * FRAMES:
        x += STEP
    else:
        y += STEP
    obj.center = (x, y)
    return obj,
```

Відобразимо анімацію:

```python
# call the animator. blit=True means only re-draw the parts that have changed.
animation.FuncAnimation(
    fig, 
    animate, 
    init_func=init,
    frames=FRAMES,
    interval=40,
    blit=True,
)
```

## Висновки

При виконанні даної лабораторної роботи я переміщував об'єкт по квадату проти годинникової стрілки.
Для цього я перші $\frac{FRAMES}{4}$ кадрів переміщував об'єкт ліворуч, наступні $\frac{FRAMES}{4}$ кадрів -- вниз, потім праворуч і вгору.

Тут $FRAMES$ -- кількість кадрів, протягом яких рухається об'єкт.

Важливими параметрами для відображення анімації є 

$frames$ -- кількість кадрів анімації
$interval$ -- затримка між відображенням кадрів у мілісекундах

Зокрема, анімація руху об'єкту по квадрату проти годинникової стрілки (у цій роботі) використовує такі параметри:
$$frames = 60$$
$$interval = 40$$
