#### Варіант 12

$$ z = \frac{(x-a)^2}{a} + \frac{(y-b)^2}{b} $$


```python
%matplotlib notebook
import numpy as np
import pandas as pd
```

Зафіксуємо значення $a$ і $b$:

$$ a = 2 $$
$$ b = 7 $$

```python
a = 2
b = 7

def z(x, y):
    return ((x - a) ** 2) / a + ((y - b) ** 2) / b
```

Побудуємо таблицю значень:

```python
# generate ranges
xs = list(range(-12, 13, 1))
ys = list(range(-6, 7, 1))

xcnt = len(xs)
ycnt = len(ys)

xs *= ycnt
ys *= xcnt

zs = [z(x, y) for x, y in zip(xs, ys)]

data_dict = {
    'xs': xs,
    'ys': ys,
    'zs': zs,
}

data = pd.DataFrame(data_dict)
data
```

За таблицею побудуємо графік.

```python
from bokeh.io import output_notebook
output_notebook()
```

```python
from bokeh.charts import Line, show

table = pd.DataFrame(index=xs[:xcnt])
for y_val, df in data.groupby(data.ys):
    table['y = {}'.format(y_val)] = list(df.sort_values(by='xs').zs)

show(Line(table, xlabel='x', ylabel='z(x,y)', title='Залежність z від x'))
```

Побудуємо карту розподілу (heatmap)

```python
from bokeh.charts import HeatMap, show
show(
    HeatMap(data_dict, x='xs', y='ys', values='zs', stat=None, title='Розподіл z залежно від x та y')
)
```

Будуємо гістограму розподілу величини $zs$ -- результату функції.

```python
from bokeh.charts import Histogram

hist = Histogram(data, values='zs', title="Розподіл результату функції")
show(hist)
```

Представимо розподіл результату у вигляді секторної діаграми

```python
from bokeh.charts import Donut

df = data.copy()
df.zs = (df.zs // 15 * 15)
df.zs = df.zs.map(lambda i: '> {}'.format(i))
df.sort_values(by='ys')

donut1 = Donut(df, label=['ys', 'zs'], color='xs', agg='count', title='Розподіл результату функції в розрізі параметра y')
donut2 = Donut(df, label='zs', agg='count', title='Загальний розподіл результату функції')
show(donut1)
show(donut2)
```
