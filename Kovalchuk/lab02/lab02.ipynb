Побудована фігура — графік кривої під назвою [Епіциклоїда](https://uk.wikipedia.org/wiki/%D0%95%D0%BF%D1%96%D1%86%D0%B8%D0%BA%D0%BB%D0%BE%D1%97%D0%B4%D0%B0) з певними параметрами.

Рівняння епіциклоїди в параметричній формі має вигляд:

$$x = r \cdot (k + 1) \cdot (cos(t) - \frac{cos((k + 1) \cdot t)}{k + 1} )$$
$$y = r \cdot (k + 1) \cdot (sin(t) - \frac{sin((k + 1) \cdot t)}{k + 1} )$$,

де $r$ — радіус кола, що рухається а $k$ — співвідношення радіусів нерухомого та рухомого кіл.

Тоді у вихідних рівняннях:
$$R = r \cdot (k + 1)$$
$$m = (k + 1)$$

Частинним є випадок, коли радіуси рухомого і нерухомого кіл однакові, тобто
$$k = 1$$ 
Тоді 
$$R = 2$$
$$m = 2$$
В цьому випадку крива називається кардіоїдою, а її графік має вигляд

```python
R = 2
m = 2
x = R * (np.cos(t) - np.cos(m * t) / m)
y = R * (np.sin(t) - np.sin(m * t) / m)
```

```python
plt.axes().set_aspect('equal', 'datalim')
lines = plt.plot(x, y, 'b')
```

Також окремо виділяють випадок, коли радіуси нерухомого і рухомого кіл співвідносяться як 1 до 2, тобто
$$k=2$$
 
Тоді
$$R=3$$
 
$$m=3$$
 
В цьому випадку крива називається нефроїдою, а її графік має вигляд

```python
R = 3
m = 3
x = R * (np.cos(t) - np.cos(m * t) / m)
y = R * (np.sin(t) - np.sin(m * t) / m)
```

```python
plt.axes().set_aspect('equal', 'datalim')
lines = plt.plot(x, y, 'b')
```
