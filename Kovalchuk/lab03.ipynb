####  Варіант 10

$$z = \frac{(x - a)^2}{a} + \frac{(y - b) ^ 2}{b}$$

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
a = 1
b = -1
```

```python
x = np.linspace(-15, 15, 1000)
y = np.linspace(-10, 10, 1000)
x, y = np.meshgrid(x,y)

def z(x, y):
    return ((x - a) ** 2) / a + ((y - b) ** 2) / b
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(x, y, z(x,y), cmap=cm.jet, linewidth=0.2)
```

# Висновок

- Отриманий графік було отримано при параметрах $a = 1$, $b = -1$
- При $ sign(a) = sign(b) $ поверхня "тягнеться" в одному напрямку, при цьому якщо $ sign(a) = sign(b) = -1 $, то усі значення $ z $ поверхні -- недодатні, при $ sign(a) = sign(b) = 1 $ усі значення $ z $ поверхні -- невід'ємні
- При $ sign(a) = -sign(b) $ значення $ z $ поверхні набуває як додатніх, так і від'ємних значень, а поверхня тягнеться у різні боки
- При $ a = 0 $ або $b = 0$ поверхня не визначена
